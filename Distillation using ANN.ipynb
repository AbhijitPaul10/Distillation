# Data Preprocessing

```python
dist.columns
```

```python
dist_XD = dist.drop("XW",axis=1)
```

```python
dist_XW = dist.drop("XD",axis=1)
```

```python
dist_XD
```

```python
dist_XW
```

```python
X_XD= dist_XD.drop("XD",axis=1)
X_XW= dist_XW.drop("XW",axis=1)
Y_XD = dist_XD.XD
Y_XW = dist_XW.XW
```

# Model building for Distillate(XD,Top product)

```python
from sklearn.model_selection import train_test_split
X_XD_train,X_XD_test,Y_XD_train ,Y_XD_test = train_test_split(X_XD,Y_XD, test_size = 0.3 , shuffle =True)
```

```python
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
X_XD_train = SC.fit_transform(X_XD_train)
X_XD_test = SC.transform(X_XD_test)
```

```python
X_XD_test.shape
```

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    keras.layers.Dense(12, input_shape=(12, ) , activation = "linear"),
    keras.layers.Dropout( rate =0.2),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(10, activation ="linear"),
    keras.layers.Dropout( rate =0.2),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(6 , activation = "linear"),
    keras.layers.Dropout( rate =0.2),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(6, activation = "linear"),
    keras.layers.Dropout( rate =0.2),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(1 , activation = "linear")
])

model.compile(optimizer = "adam" , loss = "mean_squared_error" , metrics = ["mae"])
model.values= model.fit(X_XD_train,Y_XD_train ,batch_size = 32, validation_data=(X_XD_test,Y_XD_test), epochs = 500 )
```

```python
model_history =pd.DataFrame(model.history.history)
model_history["Epochs"] = range(1,501)
model_history.rename(columns={"loss":"Loss","mae":"Mean Absolute Error","val_loss":"Validation Loss","val_mae":"Validation Mean Absolute Error"}, inplace =True)
model_history= model_history.iloc[:,[4,0,2,1,3]]
model_history
```

# Evaluating the Model

```python
plt.figure()
plt.plot(model_history["Epochs"] ,model_history["Loss"], label="Training Loss")
plt.plot(model_history["Epochs"] ,model_history["Validation Loss"], label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss for XD(Top product)")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(model_history["Epochs"] ,model_history["Mean Absolute Error"], label="Training Mean Absolute Error")
plt.plot(model_history["Epochs"] ,model_history["Validation Mean Absolute Error"], label="Validation Mean Absolute Error")
plt.xlabel("Epochs")
plt.ylabel("Mean Absolute Error")
plt.title("Mean Absolute Error for XD(Top Product)")
plt.legend()
plt.show()
```

```python
model.evaluate(X_XD_test,Y_XD_test)
```

```python
model.evaluate(X_XD_train,Y_XD_train)
```

```python
Y_XD_pred = model.predict(X_XD_test)
```

```python
from sklearn import metrics

print('Explained Variance Score:',metrics.explained_variance_score(Y_XD_test,Y_XD_pred))
print('Max Error               :',metrics.max_error(Y_XD_test,Y_XD_pred))
print('Mean Absolute Error     :',metrics.mean_absolute_error(Y_XD_test,Y_XD_pred))
print('Mean Square Error       :',metrics.mean_squared_error(Y_XD_test,Y_XD_pred))
print('Mean Squared Log Error  :',metrics.mean_squared_log_error(Y_XD_test,Y_XD_pred))
print('Median Absolute Error   :',metrics.median_absolute_error(Y_XD_test,Y_XD_pred))
print('R2 Score                :',metrics.r2_score(Y_XD_test,Y_XD_pred))
print('Mean Poisson Deviance   :',metrics.mean_poisson_deviance(Y_XD_test,Y_XD_pred))
print('Mean Gamma Deviance     :',metrics.mean_gamma_deviance(Y_XD_test,Y_XD_pred))
print('Mean Tweedie Deviance   :',metrics.mean_tweedie_deviance(Y_XD_test,Y_XD_pred))
print("Root Mean Square Error  :",np.sqrt(metrics.mean_squared_error(Y_XD_test,Y_XD_pred)))
```
